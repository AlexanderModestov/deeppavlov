```python
import re

def contains_terms(phrase, terms):
    """Check if a phrase contains any of the specified terms using regex."""
    pattern = r'\b(?:{})\b'.format('|'.join(re.escape(term) for term in terms))
    print(r'\b(?:{})'.format('|'.join(re.escape(substring) + '.*' for substring in terms)))
    pattern = r'\b(?:{})'.format('|'.join(re.escape(substring) + '.*' for substring in terms))
    return bool(re.search(pattern, phrase, re.IGNORECASE))
```
