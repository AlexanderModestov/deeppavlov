```python
import xml.etree.ElementTree as ET

def display_element(element, indent=0):
    print(' ' * indent + element.tag)
    for child in element:
        display_element(child, indent + 2)

tree = ET.parse('/content/drive/MyDrive/ColabNotebooks/SkillBox/Data/fast_track_xml_data.xml')
root = tree.getroot()

#display_element(root)
```

```python
# Iterate over all <nomenclatureTitle> tags
tittles, ids, prices, descriptions = [], [], [], []
for nomenclature in root.findall('.//offer'):
    for item in nomenclature.findall('.//nomenclatureTitle'):
        tittles.append(item.text)
    for item in nomenclature.findall('.//nomenclatureId'):
        ids.append(item.text)
    for item in nomenclature.findall('.//price'):
        prices.append(item.text)
    #print('----------------------------------------------------------------------------')
    for item in nomenclature.findall('.//skills'):
        desc = []
        for item in item:
            desc.append(item.find('description').text)
        descriptions.append(" ".join(desc))
    #print('----------------------------------------------------------------------------')
    #print('----------------------------------------------------------------------------')  # This will print the text content of the tag
```

```python
import pandas as pd
df = pd.DataFrame({'tittle': tittles, 'id':ids, 'price': prices, 'description': descriptions})
```

```python
df.to_csv('/content/drive/MyDrive/ColabNotebooks/SkillBox/Data/descriptions.csv')
```
